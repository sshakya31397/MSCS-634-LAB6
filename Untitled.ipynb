```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the dataset
df = pd.read_excel("OnlineRetail.xlsx")
print("Original data shape:", df.shape)

# Step 3: Clean the data
# Drop rows with missing invoice numbers or product descriptions
df = df[df['InvoiceNo'].notnull() & df['Description'].notnull()]

# Keep only rows with positive quantities
df = df[df['Quantity'] > 0]

# Remove cancelled transactions (InvoiceNo starting with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

print("Cleaned data shape:", df.shape)

# Optional: Check for any remaining missing values
print("Missing values:\n", df.isnull().sum())

# Step 4: Explore top-selling items
top_items = df['Description'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_items.values, y=top_items.index)
plt.title("Top 10 Most Frequently Purchased Items")
plt.xlabel("Quantity Purchased")
plt.ylabel("Item Description")
plt.tight_layout()
plt.show()

# Step 5: Create the basket (pivot table)
basket = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0)

print("Basket shape:", basket.shape)

# Display basket preview if not empty
if not basket.empty:
    print("Basket preview:")
    basket.head()
else:
    print("Basket is empty. Check earlier filters or data issues.")
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Step: Convert quantity basket to binary
basket_binary = basket.copy()
basket_binary[basket_binary > 0] = 1
basket_binary = basket_binary.astype(bool)

# Step: Apply Apriori
frequent_apriori = apriori(basket_binary, min_support=0.02, use_colnames=True)

# Display itemsets
print("Frequent itemsets found:", len(frequent_apriori))
display(frequent_apriori.head())

# Step: Plot Top 10
if not frequent_apriori.empty:
    top_apriori = frequent_apriori.sort_values('support', ascending=False).head(10).copy()
    top_apriori['itemset_str'] = top_apriori['itemsets'].apply(lambda x: ', '.join(list(x)))

    plt.figure(figsize=(10, 5))
    sns.barplot(x='support', y='itemset_str', data=top_apriori)
    plt.title("Top 10 Frequent Itemsets (Apriori)")
    plt.xlabel("Support")
    plt.ylabel("Itemset")
    plt.tight_layout()
    plt.show()
else:
    print("No frequent itemsets found.")
```

```python
from mlxtend.frequent_patterns import fpgrowth

# Step 1: Convert basket to binary (required for FP-Growth)
basket_binary = basket.copy()
basket_binary[basket_binary > 0] = 1
basket_binary = basket_binary.astype(bool)

# Step 2: Apply FP-Growth
frequent_fpgrowth = fpgrowth(basket_binary, min_support=0.02, use_colnames=True)

# Step 3: Display and visualize
print("Frequent itemsets found:", len(frequent_fpgrowth))
display(frequent_fpgrowth.head())

# Optional: plot top 10 frequent itemsets
if not frequent_fpgrowth.empty:
    top_fp = frequent_fpgrowth.sort_values('support', ascending=False).head(10).copy()
    top_fp['itemset_str'] = top_fp['itemsets'].apply(lambda x: ', '.join(list(x)))

    plt.figure(figsize=(10, 5))
    sns.barplot(x='support', y='itemset_str', data=top_fp)
    plt.title("Top 10 Frequent Itemsets (FP-Growth)")
    plt.xlabel("Support")
    plt.ylabel("Itemset")
    plt.tight_layout()
    plt.show()
else:
    print("No frequent itemsets found with FP-Growth.")
```

```python
rules_apriori = association_rules(frequent_apriori, metric="confidence", min_threshold=0.4)
rules_fpgrowth = association_rules(frequent_fpgrowth, metric="confidence", min_threshold=0.4)

# Plot confidence vs lift
plt.figure(figsize=(8, 6))
sns.scatterplot(data=rules_apriori, x='confidence', y='lift', label='Apriori')
sns.scatterplot(data=rules_fpgrowth, x='confidence', y='lift', label='FP-Growth')
plt.title("Confidence vs Lift (Association Rules)")
plt.legend()
plt.show()

# View sample rules
print(rules_apriori[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
print(rules_fpgrowth[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```
